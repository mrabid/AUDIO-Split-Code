```python
from pydub import AudioSegment
from pydub.silence import split_on_silence
from pydub.playback import play
import IPython
from scipy.io.wavfile import read
```

```python
from google.colab  import drive
drive.mount('/content/drive/')
```

```python
src='/content/drive/MyDrive/Research/Slang Final Data/Final Data-Voice Record/60/Audio/Hana_5.8ft_74kg_22y_Sylhet.wav'
```

```python
audio=AudioSegment.from_wav(src)
```

```python
rt, dt = read(src)
IPython.display.Audio(data=dt, rate=rt)
```

```python

def audio_split(minS, silanceT):
  audio_chunks = split_on_silence(audio,min_silence_len= minS, silence_thresh= silanceT)
  return audio_chunks
    # consider it silent if quieter than -16 dBFS
    # must be silent for at least half a second

#len(audio_chunks)
```

```python
minS = 250
silanceT = -32
audio_chunks = audio_split(minS, silanceT)
```

```python
print(len(audio_chunks))
```

```python
a = minS
b = silanceT
```

```python
while True:
  if len(audio_chunks) == 115 or len(audio_chunks) == 116: # or len(audio_chunks) == 117:
    finalChunks = audio_chunks
    break
  else:
    a += 10
    b -= 1 
    print(a, b)
    audio_chunks = audio_split(a, b)
    print(len(audio_chunks))

print('Final: ', len(finalChunks))
```
